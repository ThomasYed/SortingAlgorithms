```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("datagraph.txt") as file:
  lines = [line.split() for line in file]
```

```python
x = np.array(lines[0]).astype(int)
y1 = np.array(lines[1]).astype(float)
y2 = np.array(lines[2]).astype(np.float)
y3 = np.array(lines[3]).astype(np.float)
```

```python
plt.figure(figsize=(10, 8), dpi=100)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(['Bubble Sort', 'Shaker Sort', 'Merge Sort'])
plt.title("Зависимость времени сотрировки от объёма данных")
plt.show()
```

```python
plt.figure(figsize=(10, 8), dpi=100)
plt.plot(x, y3)
plt.legend(['Merge Sort'])
plt.title("Зависимость времени сотрировки Merge Sort от объёма данных")
plt.show()
```

```python
y1 = np.log(y1)
y2 = np.log(y2)
y3 = np.log(y3)
```

```python
plt.figure(figsize=(10, 8), dpi=100)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(['Bubble Sort', 'Shaker Sort', 'Merge Sort'])
plt.title("Зависимость времени сотрировки от объёма данных")
plt.show()
```
